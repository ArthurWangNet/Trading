```python
import time_utilities
import pandas as pd
import numpy as np
import datetime
import os
import pytz
import TD_API
import Paths
import time
import api_keys
import requests
import utilities
```

```python
# We first need a function which will get all eod data up to today. This function will run at the first time, and will run every time a split happens.


def get_minute_data_from_start(symbol):
	#Define payload
	payload = {
		'apikey': api_keys.APIKEY_MINUTE,
		'periodType' : 'day',
        'frequencyType': 'minute',
        'frequency':'1',
		'startDate': '21600000' #This is 1970-01-01
	}
	endpoint = TD_API.get_price_history_endpoint(symbol)

	#Get data
	try:
		time.sleep(0.6)
		td_response = requests.get(url=endpoint, params=payload)
		content = td_response.json()
		df = pd.DataFrame(content['candles'])
		#df.to_csv('./Data/AAPL.csv', index=False)
		return df
	except:
		print(symbol + " failed.")
		return symbol

```
